# K-Means

K-Means é dos principais e mais populares algoritmos de aprendizado não-supervisionado em Machine Learning.

Tudo que você viu até agora envolveu uma coluna com as variáveis alvos, um guia para o algoritmo aprender a usar as *features* e classificar ou realizar a regressão de uma amostra.

No entanto, o k-means é um modelo de *clustering* que tenta separar os seus dados em *k* *clusters*, baseando-se apenas nos padrões existentes nos dados.

## k-means no `scikit-learn`

O conjunto de dados utilizado neste notebook pode ser encontrado [neste site](https://datahack.analyticsvidhya.com/contest/practice-problem-loan-prediction-iii/?utm_source=blog&utm_medium=comprehensive-guide-k-means-clustering) ou baixado diretamente do [meu repositório no Github](https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/clustering.csv).


```python
# importar os pacotes necessários
import pandas as pd
import matplotlib.pyplot as plt

# importar o arquivo
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/clustering.csv")

# ver as primeiras entradas
print("Tamanho de df: {}\n".format(df.shape))

df.head()
```

```python
# importar scikit-learn para criar o modelo
from sklearn.preprocessing import StandardScaler
```

Nesta etapa, vamos dividir o *dataset* entre treino e teste, da mesma maneira que fizemos e sempre vamos fazer para outros modelos.

```python
# 1. escolher e importar um modelo
from sklearn.cluster import KMeans

# 2. Instanciar e escolher os hyperparameters
model = KMeans(n_clusters=2)
model2 = KMeans(n_clusters=4)

# 3. Separar os dados
X = df[['ApplicantIncome', 'LoanAmount']]

# 3.1 Padronizar os dados de treino
# scaler = StandardScaler().fit(X)
# X = scaler.transform(X)

# 4. Fit do modelo (treinar)
model.fit(X)
model2.fit(X)

# 5. Fazer previsões em cima de novos dados
y_pred = model.predict(X)
y_pred2 = model2.predict(X)
```

```python
plt.scatter('ApplicantIncome', 'LoanAmount', data=df)
plt.show()
```

```python
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred, s=40, cmap='viridis')
plt.show()
```

```python
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred2, s=40, cmap='viridis')
plt.show()
```
