# Introducción al Scikit-Learn

El proyecto `Scikit-learn` visa providenciar una biblioteca de machine learning para Python.

<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/0/05/Scikit_learn_logo_small.svg/1200px-Scikit_learn_logo_small.svg.png" width=400px ></center>

Pensada para encajar con los paquetes científicos que orbitan alrededor de las bibliotecas `Numpy` y `Scipy`, la visión de Scikit-learn es ser una herramienta eficiente y estable, al mismo tiempo que accesible para los no expertos en aprendizaje automático.

De acuerdo con su [documentación oficial](https://scikit-learn.org/stable/getting_started.html), el `Scikit-learn` es una biblioteca *open source* que soporta aprendizaje supervisado y no supervisado, además de proveer varias herramientas para *model fitting*, preprocesamiento de datos, selección de modelos y evaluación, entre otras muchas utilidades.

Para profundizar en la arquitectura de la biblioteca, recomiendo leer una de las mejores referencias que encontré, el artículo [API design for machine learning software: experiences from the scikit-learn project](https://arxiv.org/abs/1309.0238).

## Representación de datos en `Scikit-learn`

La opción del equipo de desarrollo de la biblioteca fue elegir la representación de datos que más se asemeja a la representación matricial, los *datasets* se codifican de dos maneras posibles:

1. Arrays multidimensionales de `Numpy`, para datos densos; y
2. Matrices dispersas de `Scipy`, para dados esparços.

Según el artículo referenciado anteriormente, la decisión de usar este tipo de representación, incluso habiendo otras más sofisticadas, se debió al hecho de que esto permite contar con toda la eficiencia de las operaciones vectorizadas de `Numpy` e `Scipy` - además de dejar el código claro y legible.



```python
# importando o conjunto de dados
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/heart-disease-uci.csv")
df.head()
```

Como ya sabes, cada fila representa una entrada única, un registro en la base de datos y cada columna representa una  *feature* (variable, atributo).

En el caso de este DataFrame, para crear un modelo de regresión, sería necesario dividirlo en una matriz `X` que contenga las variables independientes (*features*) y un vector objetivo `y` que contenga las variables dependientes (variables objetivo).


```python
X = df.drop('num', axis=1)
y = df.num

print("df.shape:\t{}".format(df.shape))
print("X.shape:\t{}".format(X.shape))
print("y.shape:\t{}".format(y.shape))
```

## Interfaces de `Scikit-learn`

El diseño de la biblioteca está organizado en torno a [tres APIs fundamentales](https://towardsdatascience.com/scikit-learn-design-principles-d1371958059b), pilares básicos de `Scikit-learn`: ***Estimator, Predictor*** y ***Transformer***, y es una prueba del poder de los *design patterns*.

Ten en cuenta que las interfaces no se excluyen mutuamente ni tienen un límite bien definido. De hecho, son complementarias. Vamos a conocer las tres interfaces.

### Estimator API

Son parte del corazón de `Scikit-learn`. Es a través de esta interfaz que instancias los objetos y ejecutas el método `fit` para entrenar tu modelo.

Quiero reforzar aquí que todos los algoritmos de machine learning de la biblioteca están implementados mediante la API Estimator.

[Como resumió bien el autor Jake VanderPlas](https://jakevdp.github.io/PythonDataScienceHandbook/05.02-introducing-scikit-learn.html), autor del libro Python Data Science Handbook, los pasos más comunes para usar esta API son:

1. Elegir la clase del modelo, importando la estimator class apropiada dentro de `Scikit-learn`.
2. Elegir los hiperparámetros del modelo, instanciando esta clase con los valores deseados.
3. Separar los datos en matriz de *features* y vector objetivo.
4. Ajustar el modelo, ejecutando el método `fit()` en la instancia del modelo.
5. Aplicar el modelo a los datos nuevos, lo que puede ser:
    * Para aprendizaje supervisado, comúnmente se utiliza el método `predict()`.
    * Para aprendizaje no supervisado, comúnmente transformamos o inferimos las propiedades de los datos usando `transform()` o `predict()`.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000, penalty='l2')
model.fit(X, y)
```
